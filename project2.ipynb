```python
print("Welcome to the Pattern Generator and Number Analyzer!")

while True:
    print("Select an option:")
    print("1. Generate a Pattern.")
    print("2. Analyze a Range of Number.")
    print("3. Exit.")

    n = int(input("Enter your choice :"))

    if n==1:
        n1=int(input("Enter number of rows:"))

        if n1<=0:
             print("Invalid input.Number of Rows must be a positive number.")
        else:
            for i in range(1,n1+1):
                print("*"*i)
                

    elif n==2:
        s=int(input("Enter starting number of range:"))
        e=int(input("Enter ending number of range:"))

        if s>e:
            print("Invalid range.Please enter a valid range.")
        else:
            total=0
            for i in range(s,e+1):
                if i==0:
                    pass
                if i<0:
                    continue
                if i%2==0:
                    print(f"{i} is an even number.")
                else:
                    print(f"{i} is an odd number.")
                
                total+=i
                
            print("Sum of all numbers is ",total)
    
    elif n==3:
        print("Exiting the program.")
        print("Have a good day:)")
        break
    else:
        print("Invalid choice,please try again.")
```
